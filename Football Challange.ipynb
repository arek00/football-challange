```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
pd.options.display.max_columns = None
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
def bestHome(row):
    elems = [row.B365H, row.BWH, row.IWH, row.LBH, row.WHH, row.VCH]
    return max(elems)

def bestAway(row):
    elems = [row.B365A, row.BWA, row.IWA, row.LBA, row.WHA, row.VCA]
    return max(elems)

def bestDraw(row):
    elems = [row.B365D, row.BWD, row.IWD, row.LBD, row.WHD, row.VCD]    
    return max(elems)
```

```python
df['bestHome'] = df.apply(lambda x: bestHome(x), axis = 1)
df['bestAway'] = df.apply(lambda x: bestAway(x), axis = 1)
df['bestDraw'] = df.apply(lambda x: bestDraw(x), axis = 1)
```

```python
df.head()
```

```python
df = df[['ID', 'Date', 'HomeTeam', 'AwayTeam', 'FTR', 'bestHome', 'bestAway', 'bestDraw']]
```

```python
df.head()
```

```python
def favourite(row):
    bets = [row.bestHome, row.bestAway, row.bestDraw]
    
    minBet = min(bets)
    
    if(minBet == row.bestHome): return "H"
    if(minBet == row.bestAway): return "A"
    if(minBet == row.bestDraw): return "D"
    
```

```python
df['favourite'] = df.apply(lambda x: favourite(x), axis = 1)
```

```python
df.head(10)
```

```python
df['match'] = df.apply(lambda x: (x.ID - 611) // 10, axis = 1)
df['season'] = df.apply(lambda x: x.match // 38 + 1, axis=1)
df['match_'] = df.apply(lambda x: x.match % 38, axis = 1)
```

```python
matchesCount = len(df)
favorableResult = df.query("FTR == favourite")

print("Favorable result percentage: {}".format(len(favorableResult) / matchesCount))
```

```python
def onFavouriteOutput(row, stake, minRate, maxRate):
    favouriteBet = min([row.bestHome, row.bestAway, row.bestDraw])
    
    if(favouriteBet < minRate or favouriteBet > maxRate):
        return 0
    
    if(row.FTR == row.favourite):
        return (stake * favouriteBet) - stake
    else:
        return stake * -1
```

```python
def betOnFavouriteStrategy(df, stake, minBet, maxBet):
    df['ret'] = df.apply(lambda x: onFavouriteOutput(x, stake, minBet, maxBet), axis = 1)
    return df
```

```python
stake = 1
lowestRate = 1
maxRate = 2.6
onFavouriteDf = betOnFavouriteStrategy(df, stake, lowestRate, maxRate)
```

```python
playedMoney = len(onFavouriteDf.query("ret != 0")) * stake
bettingReturn = np.sum(onFavouriteDf.ret)
print("Returning money {}. Played: {}".format(bettingReturn, playedMoney))
```
